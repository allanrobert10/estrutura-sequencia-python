```python
# QUESTÃO 15: Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. 
# Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para 
# o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê:
# a) salário bruto.
# b) quanto pagou de Imposto de Renda
# c) quanto pagou ao INSS.
# d) quanto pagou ao sindicato.
# e) o salário líquido.
# f) calcule os descontos e o salário líquido, conforme a tabela abaixo:

'''
+ Salário Bruto : R$
- IR (11%) : R$
- INSS (8%) : R$
- Sindicato (5%) : R$
= Salário Liquido : R$
Obs.: Salário Bruto - Descontos = Salário Líquido.
'''

ganho_por_hora = float(input(f"Informe o valor da hora trabalhada: "))
numero_horas_trabalhadas_dia = float(input(f"Informe o número de horas trabalhadas no dia: "))

numero_horas_trabalhadas_mes = numero_horas_trabalhadas_dia * 5 * 4
print(f"\nA quantidade de horas no mês é de {numero_horas_trabalhadas_mes} horas.")

salario_bruto = ganho_por_hora*numero_horas_trabalhadas_mes
print(f"\nO valor do seu salário no mês é de R${salario_bruto:.2f}\n")

imposto_renda = (100 - 11)/100
print(f"taxa do Imposto de Renda = {imposto_renda}")
pago_imposto_renda = (salario_bruto * imposto_renda) - salario_bruto
print(f"O Valor pago ao Imposto de Renda é de R${pago_imposto_renda:.2f}\n")

inss = (100 - 8)/100
print(f"taxa do INSS = {inss}")
pago_inss = (salario_bruto * inss) - salario_bruto
print(f"O Valor pago ao INSS é de R${pago_inss:.2f}\n")

sindicato = (100 - 5)/100
print(f"taxa do Sindicato = {sindicato}")
pago_sindicato = (salario_bruto * sindicato) - salario_bruto
print(f"O Valor pago ao Sindicato é de R${pago_sindicato:.2f}\n")

salario_liquido = salario_bruto + pago_imposto_renda + pago_inss + pago_sindicato
print(f"O valor do salário líquido é de R${salario_liquido:.2f}")
```

```python
# QUESTÃO 17: Faça um Programa para uma loja de tintas. O programa deverá pedir o tamanho em 
# metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 
# 1 litro para cada 6 metros quadrados e que a tinta é vendida em latas de 18 litros, 
# que custam R$ 80,00 ou em galões de 3,6 litros, que custam R$ 25,00.
# Informe ao usuário as quantidades de tinta a serem compradas e os respectivos preços em 3 situações:
# a) comprar apenas latas de 18 litros;
# b) comprar apenas galões de 3,6 litros;
# c) misturar latas e galões, de forma que o desperdício de tinta seja menor. Acrescente 
# 10% de folga e sempre arredonde os valores para cima, isto é, considere latas cheias.

import math

# entrada do usuário
area_pintar = float(input("Digite o tamanho em metros quadrados da área a ser pintada: "))

# cálculo da quantidade de tinta necessária
litros_tinta = area_pintar / 6
litros_tinta += litros_tinta * 0.1 # acrescentando 10% de folga

# quantidade de latas de tinta
qtd_latas = math.ceil(litros_tinta / 18)

# quantidade de galões de tinta
qtd_galoes = math.ceil(litros_tinta / 3.6)

# quantidade de latas e galões misturados
qtd_latas_misturadas = qtd_latas
qtd_galoes_misturados = 0

# ajuste para adicionar galões na mistura
if qtd_latas_misturadas > 0 and litros_tinta % 18 <= 3.6:
  qtd_latas_misturadas -= 1
  qtd_galoes_misturados = 1
elif litros_tinta % 18 > 3.6:
  qtd_galoes_misturados = math.ceil((litros_tinta % 18) / 3.6)

# cálculo do preço total
preco_latas = qtd_latas * 80
preco_galoes = qtd_galoes * 25
preco_misturado = (qtd_latas_misturadas * 80) + (qtd_galoes_misturados * 25)

# exibição dos resultados
print("\nQuantidade de tinta a ser comprada:")
print(f"{litros_tinta:.2f} litros\n")

print("Comprando apenas latas de 18 litros:")
print(f"{qtd_latas} latas de tinta")
print(f"Preço total: R$ {preco_latas:.2f}\n")

print("Comprando apenas galões de 3,6 litros:")
print(f"{qtd_galoes} galões de tinta")
print(f"Preço total: R$ {preco_galoes:.2f}\n")

print("Misturando latas e galões:")
print(f"{qtd_latas_misturadas} latas de tinta e {qtd_galoes_misturados} galões de tinta")
print(f"Preço total: R$ {preco_misturado:.2f}")
```

```python
# QUESTÃO 18: Faça um programa que peça o tamanho de um arquivo para download (em MB) 
# e a velocidade de um link de Internet (em Mbps), calcule e informe o 
# tempo aproximado de download do arquivo usando este link (em minutos).

tamanho_arquivo = float(input("Digite o tamanho do arquivo em MB: "))
velocidade_internet = float(input("Digite a velocidade da sua internet em Mbps: "))


tempo_download = (tamanho_arquivo * 8) / velocidade_internet / 60


print(f"O tempo aproximado de download é de {tempo_download:.2f} minutos.")
```
